### Task 3

```python
order_amount = []
total = 0
while True:
    print("\nMENU\n")
    print("1-Add order amount to a running list")
    print("2-Show all orders and totals after applying discounts")
    print("q-Quit")

    choice = input("Enter your choice: ")

    if choice == "1":
        price = float(input("Enter your order amount: "))
        order_amount.append(price)

    elif choice == "2":
        print(f"All the orders you have done : {order_amount}")
        for i in order_amount:
            if(i>=2000):
                total = total + (i*0.85)
            elif(i>=1500 and i<2000):
                 total = total + (i*0.9)
            elif(i>=1000 and i<1500):
                total = total + (i*0.93)
            else:
                total = total + i
        print(f"Total order amount after applying discounts : {total}")
    
    elif choice == "q":
        break

    else:
        print("Invalid choice")
```
